1. Write a program to draw a violin plot of sex against total_bill

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
sns.violinplot(x='sex', y='total_bill', data=df, hue='smoker')
plt.show()
```

2. Write a program to draw a violin plot of species against sepal length for a dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
sns.violinplot(x='species', y='sepal_length', data=df)
plt.show()
```
