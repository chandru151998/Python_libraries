1. Write a program to draw a point plot for sex against survived for a dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
  
# loading dataset
titanic_dataset = sns.load_dataset("titanic")
  
# draw pointplot
sns.pointplot(x = "sex", y = "survived", data = titanic_dataset)

# show the plot
plt.show()
```
