1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
sns.scatterplot(x='day', y='total_bill', data=df, hue='smoker')
plt.show()
```
