1. Write a program to draw box plot of life expectancy by continent for a data set

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("FiveYearData.csv")
df.boxplot(by ='continent', column =['lifeExp'], grid = False)
plt.show()
```

2. Write a program to draw a box plot of day by tips for a dataset

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("tips.csv")
df.boxplot(by ='day', column =['tip'], grid = False)
plt.show()
```
